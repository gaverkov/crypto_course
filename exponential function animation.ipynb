```python
#n = 101
#g = 7

#n = 23
#g = 5

n = 107
g = 7

def pt(i,n):
    return [cos(2*pi*i/n),sin(2*pi*i/n)]

pts = [pt(g^i,n) for i in range(1,n+2) ]

C = circle((0,0),1.05,color='red',axes=False,aspect_ratio=1)
T = sum([text(str(i),1.2*vector(pt(i,n))) for i in range(0,n)],fontsize = 'x-small',color='black')
animation = animate([ T + C + line(pts[:t],axes=False,aspect_ratio=1) for t in range(len(pts)) ] )

animation.save('expfunc_g{}_n{}.gif'.format(g,n))
```

```python
def mult_order(g,n):
    p = 1
    for i in range(n):
        p = p*g % n
        if p==1:
            return i+1
        
```

```python
n = 107
for g in range(1,n):
    print("ord({})=={}".format(g,mult_order(g,n)))
```
